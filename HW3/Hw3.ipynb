```python
class BinaryTree():

    def __init__(self,train_data,depth):
        self.left = None
        self.right = None
        self.threshold = None
        self.attribute = None
        self.train = train_data
        self.depth = depth

    def insertRight(self,train_data):
        if self.depth == 2:
            return
        if self.right == None:
            self.right = BinaryTree(train_data, self.depth+1)

    def insertLeft(self,train_data):
        if self.depth == 2:
            return
        if self.left == None:
            self.left = BinaryTree(train_data, self.depth+1)
    def growTree(self):
        if self.depth==2:
            rep={}
            maximum = 0;
            for i in self.train:
                if i['label'] in rep:
                    rep[i['label']] += 1
                else:
                    rep[i['label']] = 1
            for i in rep:
                if maximum < rep[i]:
                    maximum=rep[i]
                    self.attribute = i
                elif maximum == rep[i] and self.attribute > i:
                    self.attribute = i
            return
        left_data=[]
        right_data=[]
        allattribute=list(self.train[0].keys())
        allattribute.remove('label')
        flag = 0
        tmp=self.train[0]
        for i in self.train[1:]:
            if i['label']!=tmp['label']:
                flag=1
                break
        if flag==0:
            self.attribute=tmp['label']
            return
        mini_gini=1
        for i in allattribute:
            valuelist=[]
            kind_of_label=[]
            for j in self.train:
                valuelist.append(j[i])
                if j['label'] not in kind_of_label:
                    kind_of_label.append(j['label'])
            valuelist=list(set(valuelist))
            valuelist=sorted(valuelist)
            kind_of_label=sorted(kind_of_label)
            for j in range(len(valuelist)-1):
                valuelist[j]=(valuelist[j+1]+valuelist[j])/2
            valuelist.pop(len(valuelist)-1)
            for j in valuelist:
                num_of_kind=[[0,0] for k in range(len(kind_of_label))]
                less=0
                larger=0
                for k in self.train:
                    if k[i]<=j:
                        num_of_kind[kind_of_label.index(k['label'])][0]=num_of_kind[kind_of_label.index(k['label'])][0]+1
                        less+=1
                    else:
                        num_of_kind[kind_of_label.index(k['label'])][1]=num_of_kind[kind_of_label.index(k['label'])][1]+1
                        larger+=1
                gini_less=1
                gini_larger=1
                for k in num_of_kind:
                    gini_less-=(k[0]/less)*(k[0]/less)
                    gini_larger-=(k[1]/larger)*(k[1]/larger)
                gini=less/(less+larger)*gini_less+larger/(less+larger)*gini_larger
                if gini < mini_gini:
                    mini_gini = gini
                    self.attribute = i
                    self.threshold = j
                elif gini ==mini_gini and self.attribute>i:
                    self.attribute = i
                    self.threshold = j
                elif gini ==mini_gini and self.attribute==i and self.threshold > j:
                    self.threshold = j
        for i in self.train:
            if i[self.attribute]<=self.threshold:
                left_data.append(i)
            else:
                right_data.append(i)
        self.insertRight(right_data)
        self.insertLeft(left_data)
        self.left.growTree()
        self.right.growTree()
```
